```python
# Pairplot of original features
import matplotlib.pyplot as plt

feature_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

fig, axes = plt.subplots(len(feature_cols), len(feature_cols), figsize=(12, 10))
for i, col_i in enumerate(feature_cols):
    for j, col_j in enumerate(feature_cols):
        ax = axes[i][j]
        if i == j:
            for species, grp in df.groupby('species'):
                ax.hist(grp[col_i], alpha=0.5, color=colors[species], label=species)
        else:
            for species, grp in df.groupby('species'):
                ax.scatter(grp[col_j], grp[col_i], alpha=0.5, s=10, color=colors[species])
        if i == len(feature_cols) - 1:
            ax.set_xlabel(col_j, fontsize=7)
        if j == 0:
            ax.set_ylabel(col_i, fontsize=7)
plt.suptitle('Iris Feature Pairplot', y=1.01)
plt.tight_layout()
plt.show()
```
